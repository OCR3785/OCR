```python
df= pd.read_excel (r"C:\Users\orlik\Downloads\tasks\Bike3.xlsx")
```

```python
import statsmodels.api as sm

X = df[["weathersit", "temp", "atemp", "hum", "windspeed" ]]
y = df["registered"]

X = sm.add_constant(X)

model = sm.OLS(y, X).fit()

beta = model.params

print(beta)
```

```python
# regression plots:
import matplotlib.pyplot as plt

plt.scatter(df["temp"], df["registered"])
plt.xlabel("Temperature")
plt.ylabel("Registered")
plt.title("Registered")

x = df["temp"]
y = model.predict(sm.add_constant(x))
plt.plot(x, y, '-', color='red')

plt.show()

```
